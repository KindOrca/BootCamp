```python
import pandas as pd
import numpy as np

# url ì—ì„œ ì—‘ì…€ íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/Travel.xlsx'

# df1, df2ì— ì €ì¥
df1 = pd.read_excel(url, sheet_name = '008770 ')
df2 = pd.read_excel(url, sheet_name = '035250')

# rowê°€ ì•„ë‹Œ columnì— 'ë§¤ì¶œì•¡', 'ì˜ì—…ì´ìµ'ì´ ê°€ë„ë¡ ë°ì´í„° transposeí•˜ê¸°
df1 = df1.T
df2 = df2.T

# ì²« ë²ˆì§¸ rowë¥¼ column nameìœ¼ë¡œ ì„¤ì •
new_header = df1.iloc[0]
df1 = df1[1:]
df1.columns = new_header
new_header2 = df2.iloc[0]
df2 = df2[1:]
df2.columns = new_header2

# ë°ì´í„°ì˜ dimension í™•ì¸
df1.ndim
df2.ndim

# ê²°ì¸¡ì¹˜ ìœ„ì¹˜ í™•ì¸ í›„ 0ìœ¼ë¡œ ëŒ€ì²´
df1.isnull()
df2.isnull()
df1 = df1.fillna(0)
df2 = df2.fillna(0)

# csvë¡œ ì¶œë ¥
df1.to_csv('df1.csv')
df2.to_csv('df2.csv')
```

```python
url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/kt%26g/kt%26g_0.csv'
headers = ['ë¶„ê¸°', 'ë§¤ì¶œì•¡', 'ì˜ì—…ì´ìµ', 'ì˜ì—…ì´ìµ(ë°œí‘œê¸°ì¤€)', 'ì„¸ì „ê³„ì†ì‚¬ì—…ì´ìµ',
           'ë‹¹ê¸°ìˆœì´ìµ', 'ë‹¹ê¸°ìˆœì´ìµ(ì§€ë°°)', 'ë‹¹ê¸°ìˆœì´ìµ(ë¹„ì§€ë°°)', 'ìì‚°ì´ê³„', 'ë¶€ì±„ì´ê³„',
           'ìë³¸ì´ê³„', 'ìë³¸ì´ê³„(ì§€ë°°)', 'ìë³¸ì´ê³„(ë¹„ì§€ë°°)', 'ìë³¸ê¸ˆ', 'ì˜ì—…í™œë™í˜„ê¸ˆíë¦„',
           'íˆ¬ìí™œë™í˜„ê¸ˆíë¦„', 'ì¬ë¬´í™œë™í˜„ê¸ˆíë¦„', 'ì˜ì—…ì´ìµë¥ ', 'ìˆœì´ìµë¥ ', 'ROE(%)',
           'ROA(%)', 'ë¶€ì±„ë¹„ìœ¨', 'ìë³¸ìœ ë³´ìœ¨', 'EPS(ì›)', 'PER(ë°°)']
```

```python
# ë°ì´í„°ì…‹ ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv(url, header = None)
df.columns = headers

# 19ë…„ë„ 4ë¶„ê¸°ì˜ ë‹¹ê¸°ìˆœì´ìµ(ë¹„ì§€ë°°) ë¶€ë¶„ì„ Naë¡œ ëŒ€ì²´
df.at[2,'ë‹¹ê¸°ìˆœì´ìµ(ë¹„ì§€ë°°)'] = None

# í•´ë‹¹ ê²°ì¸¡ì¹˜ë¥¼ mean imputation ë°©ë²•ì„ ì‚¬ìš©í•˜ì—¬ ì²˜ë¦¬
df = df.fillna(df.mean())
```

## Feature Engineering

Relative Perfomance ë¼ëŠ” ìƒˆë¡œìš´ featureë¥¼ ê³„ì‚°í•˜ì„¸ìš”.

ì´ëŠ” ìµœê·¼ 1ë…„ì¹˜ ë§¤ì¶œì•¡ì˜ í‰ê· ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ 

- 10% ì´ìƒ -> `S`

ë¼ëŠ” ê°’ì„ ê°–ëŠ” featureì…ë‹ˆë‹¤.

20ë…„ë„ 2ë¶„ê¸°ì— í•´ë‹¹í•˜ëŠ” ê²°ê³¼ê°’ì€ `A`ê°€ ë‚˜ì™€ì•¼í•©ë‹ˆë‹¤.

ê°ê°ì— í•´ë‹¹í•˜ëŠ” ë“±ê¸‰ì´ ë‚˜ì˜¤ê¸° ìœ„í•´ì„œ í•„ìš”í•œ ë§¤ì¶œì•¡ì„ ì¶”ê°€ë¡œ ì„œìˆ í•˜ì„¸ìš”.

```python
# ìµœê·¼ 1ë…„ì¹˜ ë§¤ì¶œì•¡ì˜ í‰ê· ê°’
# df['ë§¤ì¶œì•¡'] = df['ë§¤ì¶œì•¡'].str.replace(',','')
df = df.astype({'ë§¤ì¶œì•¡':float})
sales_mean = df['ë§¤ì¶œì•¡'].tail(4).mean()

# cutì„ ìœ„í•œ bin, label ì„¤ì •
bin = [0, sales_mean*90/100, sales_mean*95/100, sales_mean*105/100, sales_mean*110/100, 100000000]
label = ['D','C','B','A','S']

# Relative Perfomance ë¼ëŠ” ìƒˆë¡œìš´ featureë¥¼ ê³„ì‚°
df['Relative Perfomance'] = pd.cut(df['ë§¤ì¶œì•¡'],bins = bin, labels = label)
print(df)

# ê°ê°ì— í•´ë‹¹í•˜ëŠ” ë“±ê¸‰ì´ ë‚˜ì˜¤ê¸° ìœ„í•´ì„œ í•„ìš”í•œ ë§¤ì¶œì•¡ì„ ì¶”ê°€ë¡œ ì„œìˆ 
for i in range(1, 5):
  print(label[i],':',bin[i])
```
