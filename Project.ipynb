```python
df = pd.read_csv('vgames2.csv', index_col = 'Unnamed: 0')
print(df.isnull().sum())
df.drop_duplicates()
df['Year'] = pd.to_numeric(df['Year'], errors = 'coerce')
df['NA_Sales'] = pd.to_numeric(df['NA_Sales'], errors = 'coerce')
df['EU_Sales'] = pd.to_numeric(df['EU_Sales'], errors = 'coerce')
df['JP_Sales'] = pd.to_numeric(df['JP_Sales'], errors = 'coerce')
df['Other_Sales'] = pd.to_numeric(df['Other_Sales'], errors = 'coerce')
df.dropna(inplace = True)

# column Year 변환
df['Year'] = df['Year'].astype(int)
def get_Year(x):
    if x < 50:
        year = x + 2000
    elif x < 100:
        year = x + 1900
    else:
        year = x
    return year

df['Year'] = df['Year'].apply(lambda x: get_Year(x))
df.reset_index(drop = True, inplace = True)
df
```

```python
# 지역 별 선호하는 게임 장르
df_1 = df[['Genre', 'NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
df_sum = df_1.groupby('Genre').sum()
```

```python
# 연도별 게임의 트렌드
df_2 = df[['Year','Genre', 'NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
df_2['total_Sales'] = df_2['NA_Sales'] + df_2['EU_Sales'] + df_2['JP_Sales'] + df_2['Other_Sales']
df_2 = df_2[['Year','Genre','total_Sales']]
def make_Year(x):
    if x < 1985:
        year = 1980
    elif x < 1990:
        year = 1985
    elif x < 1995:
        year = 1990
    elif x < 2000:
        year = 1995
    elif x < 2005:
        year = 2000
    elif x < 2010:
        year = 2005
    elif x < 2015:
        year = 2010
    else:
        year = 2015
    return year
df_2['Year'] = df_2['Year'].apply(lambda x: make_Year(x))

# 트렌드 분석 (장르별 판매량)
df_year_sum = df_2.sort_values('total_Sales', ascending = False)
df_year_sum = df_2.groupby(['Year','Genre'], as_index = False).sum()
df_year_sum = df_year_sum.sort_values(by = ['Year','total_Sales'], ascending = False)

df_year_sum.query('Year == 1980')

# 2015 Action 2010 Action 2005 Sports 2000 Action 1995 Role-playing 1990 Platfom 1985 Platform 1980 Shooter
#      Shooter     Shooter     Action      Sports      Sports            Fighting     Puzzle        Action

# 현재 Action 게임이 트렌드 이다.
```
