# **Part 1. 확률 이란**

## **1-1. 전체 확률의 법칙**

확률이란, 어떠한 원인에 의해 사건이 발생할때 이에 대한 결과를 예측하는 것을 말합니다.  
즉, 어떤 사건에서 어떤 일이 일어날 가능성을 **수**로 나타낸 것을 말합니다. 

확률을 이해하는데 가장 좋은 예시는 **동전 던지기** 입니다. <br>
아래에서 동전던지기 예를 통해 확률에 대해 배워보도록 합니다.

--------

동전을 던지는 행위는 **`앞면(H)`** 또는 **`뒷면(T)`**이라는 **`두개의 잠재적 결과`**를 가지고 옵니다. <br>
하지만, 동전이 공평한 동전인지 구부러진 동전인지에 따라 앞면이 나올 확률은 달라집니다. 

  1. **공평한 동전** <br>
    - 만약 동전이 공평한 동전이라면 앞면이 나올 확률은 50% 이고, 뒷면이 나올 확률 또한 50% 입니다. <br>
       ``` 
       P(H) = 0.5
       P(T) = 0.5
       ```
  2. **구부러진 동전**
    - 만약 동전의 구부러진 동전이라면, 앞면이 나올 확률은 뒷면이 나올 확률 보다 훨씬 더 커집니다.<br>
    - 만약 앞면이 나올 확률이 75%라면, 뒷면이 나올 확률은 무엇일까요?
      ```
      P(H) = 0.75
      P(T) = ?
      ```
    - 뒷면이 나올 확률은 25% 이여야 합니다. 

<br>

> **💡**  그 이유는 **전체 확률의 법칙**에 있습니다. <br>
전체 확률의 법칙이란, **`사건에 의해 발생한 모든 잠재적 결과들이 나올 확률의 합은 1이다`**로서 동전던지기는 앞면과 뒷면이라는 두개의 결과를 초래하는 사건으로, 두 사건이 일어날 확률의 합은 1이여야 합니다. <br>
  ```
  P(A) = 1 - P(^A)
  ```







## **1-2. 확률 계산**

공평한 동전과 구부러진 동전 던지기를 통해 원하는 경우의 확률를 계산해 보도록 합니다. 

확률을 계산하는 방법에는 **두가지**가 존재합니다. <br>
  - **`method_1.`** 모든 경우의 확률을 진리표를 통해 확인하고, 원하는 경우의 확률을 더해줍니다.
  - **`method_2.`** 확인하고자 하는 경우의 확률만을 곱해줍니다. 

우리는 아래의 scenario를 두가지 방법을 사용해 확률을 확인하도록 합니다.

<br>

### **a. 동전을 두번 던졌을 때의 확률** <br>
  - **공평한 동전** <br>
    - **`method_1.`** 공평한 동전을 두번 던지는 행위의 **모든 잠재적 경우**와 **각 경우의 확률**은 아래와 같습니다. 앞면과 뒷면이 각각 한번씩 나올 확률은 **2번**과 **3번**의 경우로서, 이 두 경우의 확률을 더한 0.25 + 0. 25 = 0.5이 입니다.
  > <img width='500' src='https://i.imgur.com/slAHRNH.png'> 
  
    - **`method_2.`** 앞면이 나올 확률, **`P(H)`**은 0.5이고 뒷면이 나올 확률, **`P(T)`**은 0.5 입니다. 그러므로 앞면과 뒷면이 각각 한번씩 나오는 경우의 수는 아래와 같이 두가지 경우이고, 두 경우를 더한 0.5가 앞면과 뒷면이 각각 한번씩 나올 확률 입니다. <br>  
   
      **`P(H, T) = P(H) x P(T) = 0.5 x 0.5 = 0.25`** <br>
      **`P(T, H) = P(T) x P(H) = 0.5 x 0.5 = 0.25`** <br> 

<br>

  - **구부러진 동전**
    - **`method_1.`** 앞면이 나올 확률이 0.6이고 뒷면이 나올 확률이 0.4인 구부러진 동전을 두번 던져 나올수 있는 **모든 잠재적 경우**는 아래와 같고, 두번 모두 앞면이 나올 확률은 **1번**의 경우로서, 0.36이 답입니다.
    > <img width='500' src='https://i.imgur.com/guW1vEe.png'>
    - **`method_2.`** 두번 모두 앞면이 나오는 경우 **`P(H, H)`**는 아래와 같이 0.36 입니다. <br>

      **`P(H) x P(H) = 0.6 x 0.6 = 0.36`**

<br>

### **b. 동전을 세번 던졌을 때의 확률** <br>
  - **공평한 동전**
    - **`method_1.`** 공평한 동전을 세번 던져 나올수 있는 **모든 잠재적 경우**는 아래와 같습니다. 동전을 세번 던져 앞면이 한번만 나오는 경우는 **4번, 6번, 7번** 으로서, 세 경우의 확률을 더한 0.125 + 0.125 + 0.0.125 = 0.375 이 답입니다.   
    > <img width='600' src='https://i.imgur.com/OOZLkOP.png'> <br>

    - **`method_2.`** 앞면이 하나만 나온 확률은 **`P(H, T, T)`**, **`P(T, H, T)`**, **`P(T, T, H)`** 으로서 세 확률을 더한 0.375가 답 입니다.  
    
      **`P(H, T, T) = P(H) x P(T) x P(T) = 0.5 x 0.5 x 0.5 = 0.125`** <br>
      **`P(T, H, T) = P(H) x P(T) x P(H) = 0.5 x 0.5 x 0.5 = 0.125`** <br>
      **`P(T, T, H) = P(H) x P(H) x P(T) = 0.5 x 0.5 x 0.5 = 0.125`** <br>
  
<br>

  - **구부러진 동전**
    - **`method_1.`** 구부러진 동전을 세번 던져 나올수 있는 **모든 잠재적 경우**는 아래와 같습니다. 동전을 세번 던져 앞면이 한번만 나오는 경우는 **4번, 6번, 7번** 으로서, 세 경우의 확률을 더한 0.096 + 0.096 + 0.0.096 = 0.288 이 답입니다.  
    > <img width='600' src='https://i.imgur.com/8k5Lm0W.png'> <br>
    - **`method_2.`** 세번 던져 앞면이 나올 확률은 0.288 입니다.
      
      **`P(H, T, T) = P(H) x P(T) x P(T) = 0.6 x 0.4 x 0.4 = 0.096`**<br>
      **`P(T, H, T) = P(T) x P(H) x P(T) = 0.4 x 0.6 x 0.4 = 0.096`**<br>
      **`P(T, T, H) = P(T) x P(T) x P(H) = 0.4 x 0.4 x 0.6 = 0.096`**<br>


<br>

> ⚠️ <br>
공평한 동전과 구부러진 동전 모두 각 경우의 확률을 모두 더한 값은 1 입니다. <br> 
즉, 모든 결과의 확률을 더한 값은 1로서 이는 **전체 확률의 법칙**을 따릅니다.











# **Part 2. 이항 분포 (Binomial Distribution)**

### **2-1. 이항 분포란**

<img width='550' src='https://i.imgur.com/v89L8us.png'>

우리는 위에서 하나의 동전을 여러번 던져 원하는 일련의 사건에 대한 확률을 구하는 방법에 대해 알아봤습니다. 이제 우리는 동전을 어려번 던져 **원하는 사건의 모든 잠재적 경우**와 이에 대한 **확률**을 나타내는 **이항 분포**에 대해 배워보도록 합니다.

**이항분포**란 독립적으로 반복되어지는 행위에 의해서 **`실패`** 또는 **`성공`**와 같이 **`두가지 옵션을 가지는 사건`**의 확률을 결정하는 함수이고, 동전던지기는 각 시행시 `앞면`과 `뒷면`이라는 `두가지 옵션`을 가지는 사건으로 이항 분포의 대표적인 예입니다.

위의 그래프는 동전을 16번 던져 나오는 **앞면이 나오는 모든 잠재적 경우**와 이에 대한 **확률**을 나타내는 **이항 분포** 입니다. 동전을 던지는 독립적인 행위를 여러번 시행해 앞면이 나오는 횟수가 **`X축`**이고, 앞면이 나오는 횟수에 대한 확률을 나타내는게 **`Y축`** 입니다. 위의 그래프를 통해 동전을 16번 던졌을때 앞면이 네번이 나올 확률이 가장 높은것을 알 수 있습니다.

<br>

> **💡 Tip** <br>
위의 이항 분포는 그래프가 왼쪽으로 조금 치우쳐져 있는것을 시각적으로 확인할 수 있습니다. 추후 여러분들은 **`동전 던지기의 횟수를 늘릴수록`** 위의 분포는 왼쪽이나 오른쪽으로 치우쳐 지지 않고, 중간에 위치한 **`정규분포`**를 나타냄을 배우실 겁니다.  

<br>

**이항 분포**는 결과가 **`binary한 사건`**의 모든 잠재적 경우의 수와 이에 대한 확률을 보여줍니다.<br> 그러므로, 데이터 분석에서 **소비자의 행동 패턴을 모델링** 하는데 많이 이용되어 집니다. <br>
예를 들어, 소비자의 구매를 예측하는 모델링을 할때, 소비자의 행동의 결과는 상품의 **`구매`** 또는 **`비구매`** 혹은 클릭을 눌러 **`전환`** 또는 **`비전환`** 등 초래되는 결과가 두가지인 경우가 많기에 이항 분포가 많이 사용 됩니다. <br>

그럼, 이항 분포를 만들어 나가는 과정을 아래의 **동전 던지기 사건**을 통해 배워보도록 합시다!

<br>

## **2-2. 이항 분포 계산**

우선 이항 분포의 **`X축`** 이 나타내는 것은 동전을 어려번 던졌을때, **`앞면이 나오는 횟수의 모든 잠재적 경우`** 입니다. 그리고 **`Y축`**이 나타내는 것은 **`각 경우의 확률`**로서, 모든 잠재적 경우의 확률을 더한 값은 전체 확률의 법칙에 의해 1 이여야 합니다.  

동전을 던져 나올수 있는 모든 경우의 수와 그것의 확률를 찾는 방법을 두가지 방법을 통해 배워보도록 합니다.

  - **b.** 공식을 통해 찾는 방법

그러기에 우리는 위와 같이 진리표가 아닌 **수학적 공식**을 이용해 확인 할 수 있는 방법을 아래에서 배워보도록 합니다. <br> 이 방법 또한 두가지 과정을 거쳐 계산합니다.

  - **동전을 다섯번 던진 경우** <br>

     > <img width='200' src='https://i.imgur.com/VnUjFJW.png'> <br>

   - **세개의 앞면**과 **두개의 뒷면**이 나오는 경우
     > <img width='250' src='https://i.imgur.com/pBxksEB.png'> <br>
       위와 같이 동전을 다섯번 던져 세개의 앞면과 두개의 뒷면이 나오는 경우는 **열가지**입니다. <br> `첫번째 앞면`은 다섯개의 다른 위치에서 나올수 있고, `두번째 앞면`은 네개의 다른 위치에서 나올 수 있습니다. 그리고 `세번째 앞면`은 세개의 다른 위치에서 나올 수 있음으로, 총 **`5 x 4 x 3 = 60`** 경우의 수가 있습니다. 하지만 이 경우도 중복되는 경우의 수를 피하기 위해서 이를 **`3 x 2 x 1`** 로 나누어 줍니다. 그러므로, **`60/6 = 10`**

**`n! = n x (n-1) x (n-2) x (n-3)...x 1`** <br>
예) 4! = 4 x 3 x 2 x 1

**`n`**이 동전을 던진 횟수이고 **`k`**는 앞면이 나온 경우의 수라고 했을때, **`n`**개의 동전을 던져 **`k`**개의 앞면이 나올수 있는 경우의 수를 나타내는 공식은 아래와 같습니다. <br>
```
Product of Available Positions /  Product of Duplication
```
  - **`n! / k!(n-k)!`** <br>

이 경우 앞면 세개 `HHH`가 나올 확률이 뒷면 세개 `TTT`가 나올 확률 보다  높습니다. 그러므로 각 경우의 확률을 아래의 테이블과 같이 확인합니다. <br>
`P(heads) = 0.8 and P(tails) = 0.2` <br>
  - **동전을 세번 던져 하나의 앞면이 나올 확률** <br>
    <img width='500' src='https://i.imgur.com/xIJRjzG.png'> <br>
    - 동전을 세번 던져 하나의 앞면이 나오는 경우는 세가지로 각 확률은 0.032입니다. 그러므로 총 전체 확률은 0.096입니다.   

  - **동전을 다섯번 던저 네개의 앞면이 나올 확률**

    - 각 일어나 확률과 총 일어날 경우의 수를 곱해 총 확률을 구합니다.
      0.02048 x 10 = 0.2048

  - **동전을 열두번 던저 아홉개의 앞면이 나올 확률**
   - 12!/9!(12-9)! = 220
   - 0.8^2 x 0.2^3 = 0.0011
    - 총 일어날 확률은 220 x 0.0011 = 0.236



































##**2-2. 이항 분포 Formula**

위에서 우리는 동전던지기와 같은 각 사건이 독립적이고, 두가지의 결과를 가지는 사건의 확률을 결정하는 이항 분포의 다양한 경우와 그 확률을 확인해 보았습니다. 그리고 각 사건의 확률을 도출하는데 일관적인 공식을 배워보도록 합니다.

동전을 **`n번`** 던져 앞면이 **`k번`** 나올 확률 <br>
  1. n번 동전을 던저 앞면이 k번 나오는 경우의 수: $n! / k!(n-k)!$
  2. k번 앞면이 나오고 n-k번 뒷면이 나오는 확률: ${p^k}{(1-p)^{(n-k)}}$
  3. 이항 분포의 Formula: <br>
  **`n번 동전을 던저 앞면이 k번 나오는 경우의 수`** x **`k번 앞면이 나오고 n-k번 뒷면이 나오는 확률`** <br>
  - $n! / k!(n-k)!$ x ${p^k}{(1-p)^{(n-k)}}$


# **Part 3. 조건부 확률 (Conditional Distribution)**

## **3-1. 조건부 확률이란**
일상적인 상황에서 사건들은 서로 연관되어 있습니다. 조건부 확률이란 다른 사건의 결과에 의해 영향을 받는 한 사건의 결과에 대한 확률 입니다. 

> **P(A|B)** 는 B라는 사건이 전제했을때, A의 확률을 말합니다. 

<img width='300' src='https://i.imgur.com/teH9ho6.png'>

위에서 동전던지기의 시나리오로 조건부확률에 대해 배워보도록 합니다. <br>
우리는 두 case를 통해서 조건부 확률에 대해 배워보도록 합니다. <br>

  - **CASE 1** 보자기에서 동전을 꺼내 동전을 한번 던져 앞면이 나올 확률 <br>
  - **CASE 2** 보자기에서 동전을 꺼내 동전을 두번 던져 앞면과 뒷면이 나올 확률 <br>

<br>

### **CASE 1**


하나의 보자기에 `공평한 동전`과 `구부러진 동전` 두개의 동전이 있습니다. <br>
공평한 동전을 던져 앞면이 나올 확률은 0.5 이고, 구부러진 동전을 던져 앞면이 나올 확률은 0.9 입니다.

우선, 이 가방에서 동전 하나를 꺼냅니다. 공평한 동전과 구부러진 동전이 pick 될 확률은 각 0.5로서 같습니다. <br>
그럼 동전 하나를 꺼내 동전을 던졌을때, 이 동전의 앞면이 나올 확률은 몇일까요?

이 경우 동전을 던져 앞면이 나올 확률은 동전을 던지기 사건 전인 보자기에서 어떠한 동전을 꺼냈는지에 따라 확률이 달라지고 그러므로 우리는 조건부 확률을 구해야 합니다.

<img width='700' src='https://i.imgur.com/rr14adF.png'>

위의 테이블에서 동전이 앞면이 나오는 경우는 **1**과 **3**으로 두 확률을 합치면 0.7입니다. <br>
이 확률이 두 동전이 보자기에 있을때 앞면을 꺼낼 확률로서, 만약 두개의 공평한 동전이 있을때 동전을 꺼내 앞면이 나올 확률이 **`0.5`** 일떄에 비해 앞면을 꺼낼 확률이 더 높아진 것을 알 수있습니다. 

위의 시나리오를 공식으로 표현해보겠습니다. <br>
**`P(앞면|공평한동전)P(공평한동전) + P(앞면|구부러진동전)P(구부러진동전)`** <br>
**`= 0.5 x 0.5 + 0.9 x 0.5 = 0.7`**

<br>

### **CASE 2**

이번 case 에서 우리는 보자기에서 동전을 꺼내 동전을 두번 던집니다. <br> 
첫번째 던지기에서는 앞면이 나오고, 두번째 던지기 에서는 뒷면이 나올 확률은 몇인가요? 

<img width='700' src='https://i.imgur.com/HQ8ptx1.png'>
  
위에서 경우 **2**와 **6**이 해당됩니다. 그리고 두 확률을 더한 0.125 + 0.045 = 0.17 이 입니다. <br>
**`P(앞면|공평한동전)P(뒷면|공평한동전)P(공평한동전) + P(앞면|구부러진동전)P(뒷면|구부러진동전)P(구부러진동전)`** <br>
**`= 0.5 x 0.5 x 0.5 + 0.9 x 0.1 x 0.5 = 0.125 + 0.045 = 0.17`**

























## **3-2. 조건부 확률의 예**

위에서 우리는 동전을 보자기에서 꺼내 던져 나오는 확률을 통해 조건부 확률을 배웠습니다. <br>
이제 우리는 암진단 테스트를 이용해 조건부 확률을 배워보도록 합니다. <br>

암진단 테스트의 결과가 positive로 또는 negative로 나오는 결과는, ***환자가 암이 있는지 없는지에 따라 달라집니다***. 그러기에 이도 대표적인 ***조건부 확률의 예***입니다.  

<br>

> **시나리오** <br>
전체 population의 10%가 암에 걸린 환자입니다. 그리고 암에 걸린 환자중, 90%가 암진단 테스트에서 암이 있다고 (Positive) 나왔습니다. 그리고 암이 없는 사람들중, 80%가 암진단 테스트에서 암이 없다고 (Negative) 나왔습니다. 그러므로 우리는 아래와 같이 요약할수 있습니다.
- **`P(C)`** = 0.1
- **`P(Pos|C)`** = 0.9
- **`P(Neg|^C)`** = 0.8

> - **`P(C)`**란 암에 걸리 확률을 말합니다.
- **`P(Pos|C)`**란 암에 걸린 환자중 테스트를 통해 암에 걸렸다고 나올 확률을 말합니다.
- **`P(Neg|^C)`**란 암에 걸리지 않은 환자중 테스트를 통헤 암에 걸리지 않았다고 나올 확률을 말합니다. <br>

<br>

위의 시나리오에서 전체 확률의 법칙에 의해 우리는 아래의 사항 또한 유추 할 수 있습니다.

- **`P(^C)`** = 0.9
- **`P(Neg|C)`** = 0.1
- **`P(Pos|^C)`** = 0.2



<br>

위의 정보들을 아래의 테이블에서 정리해보도록 합니다. 

> <img width='300' src='https://i.imgur.com/tP2QWiD.png'>

<br>

위의 시나리오의 데이터를 통해 아래의 질문에 대답해 보도록 합니다.

  - **Q1.** 암이 있고, 암진단 테스트에서 Positive로 나올 확률은 무엇인가요? <br>
    **P(C $\cap$ Pos) = P(Pos|C)P(C) = 0.1 x 0.9 = 0.09** <br>

  - **Q2.** 암이 있고, 암진단 테스트에서 Negative로 나올 확률은 무엇인가요? <br>
    **P(C $\cap$ Neg) = P(Neg|C)P(C) = 0.1 x 0.1 = 0.01** <br>

  - **Q3.** 암이 없고, 암진단 테스트에서 Positive로 나올 확률은 무엇인가요? <br>
    **P(^C $\cap$ Pos) = P(Pos|^C)P(^C) 0.9 x 0.2 = 0.18** <br>

  - **Q4.** 암이 없고, 암진단 테스트에서 Negative로 나올 확률은 무엇인가요? <br>
    **P(^C $\cap$ Neg) = P(Neg|^C)P(^C) = 0.8 x 0.9 = 0.72** <br>
  
  - **Q5.** 암이 있던 없던, 전체 population에서 암진단에서 암이 있다고 나올 확률은 무엇인가요? 

    **P(Pos) = P(C)P(Pos|C) + P(^C)P(Pos|^C) = 0.1 x 0.9 + 0.9 x 0.2 = 0.27**

<br>

위에서 찾은 conditional probabilities를 통해, 추후 아래의 공식을 찾아 보도록 합니다. 참고로 아래의 공식이 나타내느 것이 Baysian Theorem 입니다.
- **P(C|Pos) = P(C $\cap$  Pos) / P(Pos)** <br>



## **3-3. 조건부 확률 Formula**

**조건부 확률**에서의 요점은, <br>
한 사람에게 질병이 있는지 테스트를 통해 나온 결과는 이 사람에게 정말로 질병이 있는지 없는지에 따라 달라지는 사건이다 라는 것입니다. <br>
그러므로 조건부 확률은 동전던지기와 같이 독립적인 시행이 아닌 서로 관련되어 있는 사건에 사용됩니다. <br>

예를 들어, 의료 테스트에서의 결과는 테스트 대상자가 질병의 유무에 의해 달라지고, 이는 우리가 확률적 추론을 할때 고려해야할 사항입니다. 그 이유는, 테스트에서 환자가 Positive한 결과를 받는 확률은, 그 사람의 몸 건강상태에 따라 달라지기 때문입니다. 만약 환자에게 질병이 있다면, 테스트 결과가 positive가 나올 가능성이 높아집니다. <br>

질병의 유무에 상관없이 **`P(Test) 결과의 확률`**은 아래와 같습니다. <br>
- **`P(Test) = P(Test|Disease)P(Disease) + P(Test|^Disease)P(^Disease)`**

<br>

우리는 A와 B라는 두 사건의 조건부 확률을 아래와 같이 공식화 할 수 있습니다. <br>
- **P(A|B) = P(A $\cap$  B) / P(B)** <br>

<br>

즉, 한 사람이 질병이 있고 테스트에서 질병이 있다고 나올 확률은 아래와 같습니다. <br>
- **P(Positive|Disease) = P(Positive $\cap$ Disease) / P(Disease)** 
- **|** 는 **`given`** 의 의미이고 **$\cap$** 는 **`and`**의 의미 입니다. 

즉, 어떠한 사람이 질병이 있다는 전제 하에 테스트에서 positive로 나올 확률은, 어떠한 사람이 **`질병이 있고 테스트에서 postive로 나올 확률`**을 **`질병을 가질 확률`**로 나눈 것과 같습니다.




# **Part 4. Bayesian Theorem**

위에서 배운 조건부 확률을 베이지안 이론으로 확장시켜보도록 합니다.

Bayesian Theorem의 공식은 아래와 같습니다.

- **P(A|B) = P(A $\cap$  B) / P(B) = P(B|A) x P(A) / P(B)**

<br>

Bayesian Theorem을 위에서 배운 암을 진단하는 테스트에 적용하여 이해해보도록 합니다.

> **시나리오** <br>
전체 인구중 1% 정도가 걸리는 암이 있습니다. 암이 있는 사람중 90% 는 테스트에서 **`positive`**하게 나옵니다. 그리고 암이 없는 사람중 90%는 테스트에서 **`Negative`**하게 나옵니다.
>
>- **`P(C)`** = 0.01
- **`P(Pos|C)`** = 0.9
- **`P(Neg|^C)`** = 0.9

위의 내용에서 아래를 유추할 수 있습니다.

- **`P(^C)`** = 0.99
- **`P(Neg|C)`** = 0.1
- **`P(Pos|^C)`** = 0.1

하지만, 만약 어떤 사람이 테스트에서 **`positive`**라고 나온 상황이면, 그 사람이 암이 있을 확률은 1%보다 높아집니다. 즉, **P(Cancer|Positive)**는 1%보다 높을것 입니다. 얼만큼 높아질까요?
